```python
from Survival.Utils import load_val_data
from Survival.Utils import calc_scores
from Survival.Utils import load_whole_data
from Survival.Utils import load_score_containers

from Survival.KNNKaplanMeier import KNNKaplanMeier
# from Survival.CoxPHModel import CoxPHModel
# from Survival.AalenAdditiveModel import AalenAdditiveModel
# from Survival.WeibullRegressionModel import WeibullRegressionModel
# from Survival.RandomSurvivalForest import RandomSurvivalForest

import numpy as np
# import pickle

import matplotlib.pyplot as plt
```

```python
dataset_idxs = [0, 1] # 0: "pancreatitis", 1: "ich", 2: "sepsis"
train_dfs, test_dfs, unique_times, dataset_names = load_val_data(dataset_idxs, verbose=True, data_path="../../dataset/")
```

```python
n_neighbors = [5, 10, 15, 20, 30]
pca_flags = [True, False]
```

```python
for pca_flag in pca_flags:
    for idx, dataset_name in enumerate(dataset_names):
        sorted_unique_times = unique_times[dataset_name]
        
        filename = filename_generator("KNN", pca_flag, [dataset_idxs[idx]])
        concordances, ipecs = load_score_containers([dataset_name], [n_neighbors])
        ipecs = {dataset_name: []}
        print("\nFor the " + dataset_name + " dataset:")
        print(filename)

        for row, n_neighbor in enumerate(n_neighbors):
            print("[LOG] n_neighbor = {}".format(n_neighbor))

            tmp_concordances = []
            tmp_ipecs = []

            for index, cur_train in enumerate(train_dfs[dataset_name]):
                cur_test = test_dfs[dataset_name][index]
                model = KNNKaplanMeier(n_neighbors=n_neighbor, 
                    pca_flag=pca_flag, n_components=20)
                model.fit(cur_train, duration_col='LOS', event_col='OUT')
                concordance, ipec_score_list = \
                    calc_scores(model, cur_test, sorted_unique_times,
                                print_result=False, verbose_calc_time=False)

                tmp_concordances.append(concordance)
                tmp_ipecs.append(ipec_score_list)

            avg_concordance = np.average(tmp_concordances)
            avg_ipec = np.average(tmp_ipecs, axis=0)

            print("[LOG] avg. concordance:", avg_concordance)
            print("[LOG] avg. ipec:", 
                  avg_ipec[int(len(avg_ipec) * 0.8)], 
                  avg_ipec[int(len(avg_ipec) * 0.8)] / sorted_unique_times[int(len(avg_ipec) * 0.8)])

            concordances[dataset_name][row] = avg_concordance
            ipecs[dataset_name].append(avg_ipec)

            print("-------------------------------------------------------")

            with open(filename, 'wb') as f:
                pickle.dump([n_neighbors, concordances, ipecs], f, pickle.HIGHEST_PROTOCOL)
```

### Survival Curves

Compare the model's performance (survival function) between KNN and COX, using the ICH dataset's first val dataset. Here COX is used as the control group. For KNN, test two n_neighbor values: 30, which yields the best model on concordance, and 160, which yields the best model on IPEC.


```python
knn_model_30 = KNNKaplanMeier(n_neighbors=30)
knn_model_160 = KNNKaplanMeier(n_neighbors=160)
cox_model = CoxPHModel(alpha=1, lambda_=0.05)
```

Use the first val dataset of the ICH dataset.

```python
train_df = train_dfs["ich"][2]
test_df = test_dfs["ich"][2]
checkpoint_times = unique_times["ich"]
```

Fit the three models.

```python
knn_model_30.fit(train_df, duration_col='LOS', event_col='OUT')
knn_model_160.fit(train_df, duration_col='LOS', event_col='OUT')
cox_model.fit(train_df, duration_col='LOS', event_col='OUT')
```

Predict the survival matrices for each model.

```python
concordance, ipec_score_list = calc_scores(knn_model_30, test_df, checkpoint_times,
                                           print_result=False, verbose_calc_time=False)
print("KNN - 30:", concordance, ipec_score_list[int(len(checkpoint_times) * 0.8)])
concordance, ipec_score_list = calc_scores(knn_model_160, test_df, checkpoint_times,
                                           print_result=False, verbose_calc_time=False)
print("KNN - 160:", concordance, ipec_score_list[int(len(checkpoint_times) * 0.8)])
concordance, ipec_score_list = calc_scores(cox_model, test_df, checkpoint_times,
                                           print_result=False, verbose_calc_time=False)
print("COX:", concordance, ipec_score_list[int(len(checkpoint_times) * 0.8)])
```

```python
proba_matrix_knn30 = knn_model_30.pred_proba(test_df, checkpoint_times)
proba_matrix_knn160 = knn_model_160.pred_proba(test_df, checkpoint_times)
proba_matrix_cox = cox_model.pred_proba(test_df, checkpoint_times)
```

Plot the survival curve for random pick test datapoint

```python
datapoint_idx = np.random.randint(proba_matrix_cox.shape[0])
plt.plot(checkpoint_times, proba_matrix_knn30[datapoint_idx], label="KNN-30")
plt.plot(checkpoint_times, proba_matrix_knn160[datapoint_idx], label="KNN-160")
plt.plot(checkpoint_times, proba_matrix_cox[datapoint_idx], label="COX")
plt.title("Survival Curve for #" + str(datapoint_idx))
plt.legend()
plt.show
```
